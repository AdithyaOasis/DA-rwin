```python
import pandas as pd
```

```python
df = pd.read_csv("./data/WithLyrics.csv")
```

```python
df.head(20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
```

```python
Lyrics = df['Lyrics']
ProperLyrics = []
def isNaN(string):
    return string != string
```

```python
import numpy as np
for i in range(len(Lyrics)):
    if (isNaN(Lyrics[i])) != 1:
        ProperLyrics.append(i)
```

```python
ProperLyrics
```

```python
PropLyr = []
for i in ProperLyrics:
    PropLyr.append(Lyrics[i])
```

```python
dtm = tfidf.fit_transform(PropLyr)
```

```python
dtm
```

```python
from sklearn.decomposition import NMF
```

```python
nmf_model = NMF(n_components=20,random_state=42)
```

```python
nmf_model.fit(dtm)
```

```python
for index,topic in enumerate(nmf_model.components_):
    print(f'THE TOP 15 WORDS FOR TOPIC #{index}')
    print([tfidf.get_feature_names()[i] for i in topic.argsort()[-15:]])
    print('\n')
```

```python
topic_results = nmf_model.transform(dtm)
```

```python
topic_results.argmax(axis=1)

Topics = topic_results.argmax(axis=1)
```

```python
Topics
```

```python
len(Topics)
```

```python
FinalTopics = []
count = 0
for i in range(len(df['Lyrics'])):
    if i in ProperLyrics:
        FinalTopics.append(Topics[count])
        count = count + 1
    else:
        FinalTopics.append(-1)
```

```python
len(FinalTopics)
```

```python
df['Song Topics'] = FinalTopics
```

```python
df.head(100)
```

```python
numbers = []
for i in range(21):
    numbers.append(0)
for i in df['Song Topics']:
    numbers[i+1] = numbers[i+1] + 1
```

```python
numbers
```

```python
import matplotlib.pyplot as plt
```

```python
frequencies = []
for i in range(1,len(numbers)):
    frequencies.append(numbers[i])
```

```python
frequencies
```

```python
plt.plot(frequencies)
```

```python
indexes = []
for i in range(len(frequencies)):
    indexes.append(i)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(indexes,frequencies)
plt.xticks(np.arange(min(indexes), max(indexes)+1, 1.0))
plt.yticks(np.arange(min(frequencies), max(frequencies)+1, 2.0))
plt.xlabel('Topic Number')
plt.ylabel('Number of songs in each Topic')
plt.title('Song ditribution with NMF method')
plt.show()
```
